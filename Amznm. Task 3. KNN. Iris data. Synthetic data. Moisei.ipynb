#### Amznm. Knn. Iris Data

#### Load iris data

```python
import pandas as pd 
import numpy as np 
np.random.seed = 2021
from sklearn.datasets import load_iris
iris = load_iris()
print ('data contains:',iris.keys())
X, y, labels, feature_names  = iris.data, iris.target, iris.target_names, iris['feature_names']
df_iris= pd.DataFrame(X, columns= feature_names) 
df_iris['label'] =  y
features_dict = {k:v for k,v in  enumerate(labels)}
df_iris['label_names'] = df_iris.label.apply(lambda x: features_dict[x])
df_iris
```

#### Split to train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(iris.data)
```

```python
len(iris.data)
```

```python
print('len(X_train)={:,}'.format(len(X_train)))
print('len(X_test)={:,}'.format(len(X_train)))
```

#### Showing the training points

```python
import matplotlib.pyplot as plt
from sklearn import datasets

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.title('The training points')
plt.xticks(())
plt.yticks(())
```

#### Train KNN classifiers for different K

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
print('score train = {:.3f}'.format(knn.score(X_train,y_train)))
print('score test = {:.3f}'.format(knn.score(X_test,y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train, y_train)
print('score train = {:.3f}'.format(knn.score(X_train,y_train)))
print('score test = {:.3f}'.format(knn.score(X_test,y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors = 26)
knn.fit(X_train, y_train)
print('score train = {:.3f}'.format(knn.score(X_train,y_train)))
print('score test = {:.3f}'.format(knn.score(X_test,y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors = 33)
knn.fit(X_train, y_train)
print('score train = {:.3f}'.format(knn.score(X_train,y_train)))
print('score test = {:.3f}'.format(knn.score(X_test,y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors = 51)
knn.fit(X_train, y_train)
print('score train = {:.3f}'.format(knn.score(X_train,y_train)))
print('score test = {:.3f}'.format(knn.score(X_test,y_test)))
```

#### Locating better values of K for higher accuracy

```python
import matplotlib.pyplot as plt
%matplotlib inline
k_range= range (1,120)
scores=[]
for k in k_range :
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test,y_pred))
    
plt.title('Better values of K for higher accuracy = {}'.format(k_range))
plt.scatter(k_range,scores)
```

#### Showing how sensitive is k-NN classification accuracy to the choice of the 'k' parameter

```python
from sklearn.neighbors import KNeighborsRegressor

fig, ((ax1,ax2)) = plt.subplots(nrows=2, ncols=1, sharey=True)
k_range = range(1,120)
scores_train = []
scores_test=[]

for k in k_range:
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores_train.append(knn.score(X_train, y_train))
    scores_test.append(knn.score(X_test, y_test))


plt.sca(ax1)
plt.ylabel('accuracy')

ax1.scatter(k_range, scores_train)


plt.sca(ax2)
plt.xlabel('k')

plt.ylabel('accuracy')
ax2.scatter(k_range, scores_test)
plt.xticks([0,20,40,60,80,100,120]);
fig.subplots_adjust(hspace=.4)
ax1.set_title('train set');
ax2.set_title('test set');
```

#### Choosing the best K determined on test set

#### As we can see, on picture picture where are better values of K for higher accuracy and on our test set, best K value will be K = 33. It is because our training has better score and more important our test score has better value. If we put bigger number of K, then values of score will be worse and worse.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors = 33)
knn.fit(X_train, y_train)
print('The best training score of K = {:.3f}'.format(knn.score(X_train,y_train)))
print('Best test score = {:.3f}'.format(knn.score(X_test,y_test)))
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier

# generate dataset
X_D2, y_D2 = make_blobs(n_samples = 300, n_features = 2, centers = 8, cluster_std = 1.3, random_state = 2021)
# define bounds of the domain
min1, max1 = X_D2[:, 0].min()-1, X_D2[:, 0].max()+1
min2, max2 = X_D2[:, 1].min()-1, X_D2[:, 1].max()+1
# define the x and y scale
x1grid = np.arange(min1, max1, 0.1)
x2grid = np.arange(min2, max2, 0.1)
# create all of the lines and rows of the grid
xx, yy = np.meshgrid(x1grid, x2grid)
# flatten each grid to a vector
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
# horizontal stack vectors to create x1,x2 input for the model
grid = np.hstack((r1,r2))
# define the model
model = KNeighborsClassifier(n_neighbors = 3)
# fit the model
model.fit(X_D2, y_D2)
# make predictions for the grid
yhat = model.predict(grid)
# reshape the predictions back into a grid
zz = yhat.reshape(xx.shape)
# plot the grid of x, y and z values as a surface
plt.contourf(xx, yy, zz, cmap='Paired')
# create scatter plot for samples from each class
for class_value in range(225):
    # get row indexes for samples with this class
    row_ix = np.where(y_D2 == class_value)
    # create title of these plot
    plt.title('KNN Classifier = {}'.format(model))
    # create scatter of these samples
    plt.scatter(X_D2[row_ix, 0], X_D2[row_ix, 1], cmap='Paired')
```
